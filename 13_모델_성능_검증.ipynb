## 1. лҚ°мқҙн„°мқҳ нҷ•мқёкіј мҳҲмёЎ мӢӨн–ү

```python
import pandas as pd

# к№ғн—ҲлёҢм—җ мӨҖл№„лҗң лҚ°мқҙн„° нҳём¶ң
!git clone https://github.com/taehojo/data.git

# кҙ‘л¬ј лҚ°мқҙн„° нҳём¶ң
df=pd.read_csv('./data/sonar3.csv',header=None)

# мІ« 5мӨ„
df.head()
```

```python
# мқјл°ҳ м•”м„қ(0)кіј кҙ‘м„қ(1)мқҳ к°ңмҲҳ нҷ•мқё
df[60].value_counts()
```

```python
# X: мқҢнҢҢ кҙҖл Ё мҶҚм„ұ, y: кҙ‘л¬јмқҳ мў…лҘҳ
X=df.iloc[:,0:60]
y=df.iloc[:,60]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# лӘЁлҚё м„Өм •
model=Sequential()
model.add(Dense(24,input_dim=60,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

# лӘЁлҚё м»ҙнҢҢмқј
model.compile(loss='binary_crossentropy',optimizer=
              'adam',metrics=['accuracy'])

# лӘЁлҚё мӢӨн–ү
history=model.fit(X,y,epochs=200,batch_size=10)
```

## 3. н•ҷмҠөм…Ӣкіј н…ҢмҠӨнҠём…Ӣ

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split

import pandas as pd
```

```python
# к№ғн—ҲлёҢм—җ мӨҖл№„лҗң лҚ°мқҙн„° нҳём¶ң
!git clone https://github.com/taehojo/data.git
```

```python
# кҙ‘л¬ј лҚ°мқҙн„° нҳём¶ң
df=pd.read_csv('./data/sonar3.csv',header=None)
```

```python
# X: мқҢнҢҢ кҙҖл Ё мҶҚм„ұ, y: кҙ‘л¬јмқҳ мў…лҘҳ
X=df.iloc[:,0:60]
y=df.iloc[:,60]
```

```python
# н•ҷмҠөм…Ӣкіј н…ҢмҠӨнҠём…Ӣ кө¬л¶„
X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.3,shuffle=True)
```

```python
# лӘЁлҚё м„Өм •
model=Sequential()
model.add(Dense(24,input_dim=60,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

# лӘЁлҚё м»ҙнҢҢмқј
model.compile(loss='binary_crossentropy',optimizer=
              'adam',metrics=['accuracy'])

# лӘЁлҚё мӢӨн–ү
history=model.fit(X_train,y_train,epochs=200,batch_size=10)
```

```python
# лӘЁлҚёмқ„ н…ҢмҠӨнҠём…Ӣм—җ м Ғмҡ©н•ҙ м •нҷ•лҸ„ кө¬н•Ё
score=model.evaluate(X_test,y_test)
print('Test accuracy:',score[1])
```

## 5. kкІ№ кұ°м°Ё кІҖмҰқ

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

import pandas as pd

# к№ғн—ҲлёҢм—җ мӨҖл№„лҗң лҚ°мқҙн„° нҳём¶ң
!git clone https://github.com/taehojo/data.git

# кҙ‘л¬ј лҚ°мқҙн„° нҳём¶ң
df=pd.read_csv('./data/sonar3.csv',header=None)

# X: мқҢнҢҢ кҙҖл Ё мҶҚм„ұ, y: кҙ‘л¬јмқҳ мў…лҘҳ
X=df.iloc[:,0:60]
y=df.iloc[:,60]
```

```python
# лӘҮ кІ№мңјлЎң лӮҳлҲҢм§Җ м •н•Ё
k=5

# KFold н•ЁмҲҳ нҳём¶ң
kfold=KFold(n_splits=k,shuffle=True)

# м •нҷ•лҸ„к°Җ мұ„мӣҢм§Ҳ лҰ¬мҠӨнҠё мғқм„ұ
acc_score=[]

def model_fn():
  model=Sequential()      # л”Ҙлҹ¬лӢқ лӘЁлҚё кө¬мЎ° мӢңмһ‘
  model.add(Dense(24,input_dim=60,activation='relu'))
  model.add(Dense(10,activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  return model

# KкІ№ көҗм°Ё кІҖмҰқмқ„ мқҙмҡ©н•ҙ kлІҲ н•ҷмҠө мӢӨн–ү
for train_index,test_index in kfold.split(X):     # forл¬ём—җ мқҳн•ҙ kлІҲ л°ҳліө, split()м—җ мқҳн•ҙ kк°ң н•ҷмҠөм…Ӣ, н…ҢмҠӨнҠём…ӢмңјлЎң л¶„лҰ¬
  X_train,X_test=X.iloc[train_index,:],X.iloc[test_index,:]
  y_train,y_test=y.iloc[train_index],y.iloc[test_index]

  model=model_fn()
  model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
  history=model.fit(X_train,y_train,epochs=200,batch_size=10,verbose=0)

  accuracy=model.evaluate(X_test,y_test)[1]     # м •нҷ•лҸ„
  acc_score.append(accuracy)      # м •нҷ•лҸ„ лҰ¬мҠӨнҠём—җ м ҖмһҘ

# kлІҲ мӢӨмӢңлҗң м •нҷ•лҸ„ нҸүк· 
avg_acc_score=sum(acc_score)/k

# кІ°кіј м¶ңл Ҙ
print('м •нҷ•лҸ„: ',acc_score)
print('м •нҷ•лҸ„ нҸүк· : ',avg_acc_score)
```
